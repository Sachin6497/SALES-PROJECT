1. CATEGORICAL PLOTS (Seaborn)
These plots are used when the x-axis is a category (e.g., gender, product type, city, store, brand).

‚≠ê
Why Use Categorical Plots?
To compare categories:
1. Which category sells more?
2. What is average sales by store?
3. Which group has highest profit?
4. Compare distribution across categories

üìå
1.1 Countplot
Shows count of each category.
‚úî
 Use Case:
Count of customers by gender
Count of products by category
Count of stores by city

```python
import seaborn as sns 
#df= sns.load_dataset("anscombe")
d1 = sns.load_dataset("tips")
d1
```

1.1 Countplot
Shows count of each category.
‚úî
Use Case:
Count of customers by gender
Count of products by category
Count of stores by city

```python
sns.countplot(x='total_bill', data=d1)
```

This countplot shows the number of occurrences of each total¬†bill¬†amount.

üìå
1.2 Barplot
Shows mean value for each category.
‚úî
Use Case:
Average sales per store
Average rating per product



```python
sns.barplot(x='tip', y='smoker', data=d1)
```

This bar plot represents the average tip amount given by smokers and non-smokers.
The graph shows that smokers tend to give a slightly higher mean tip compared¬†to¬†non-smokers.

üìå
1.3 Boxplot
Shows distribution with median, quartiles, outliers.
‚úî
Use Case:
Salary distribution by department
Marks distribution by class


```python
sns.boxplot(x='size', y='day', data=d1)
```

This box plot shows the distribution of table sizes for different days of the week.
It compares the median, spread, and outliers of group sizes¬†on¬†each¬†day.

üìå
1.4 Violin Plot
Combination of boxplot ÓÇù KDE density.
‚úî
Use Case:
Understanding skewness and density
Comparing distributions of categories


```python
sns.violinplot(x='sex', y='time', data=d1)
```

This violin plot represents the distribution and density of meal times (Lunch and Dinner) for male and female customers.
It shows that Dinner is more frequent than Lunch for¬†both¬†genders.

üìå
1.5 Stripplot
Shows each data point on categories.



```python
sns.stripplot(x='smoker', y='sex', data=d1)
```

This strip plot represents the distribution of customers by gender and smoking status.
It shows that both males and females are present among smokers and non-smokers, with more customers in the non-smoker category.

üìå
1.6 Swarmplot
Like stripplot but no overlap.

```python
sns.swarmplot(x='total_bill', y='sex', data=d1)
```

This swarm plot represents the distribution of total bill amounts for male and female customers.
It shows individual observations without overlap, making it easy to compare spending patterns¬†by¬†gender.


1.7 Pointplot
Shows mean + confidence interval.



```python
sns.pointplot(x='tip', y='sex', data=d1)
```

This point plot represents the mean tip amount for male and female customers along with confidence intervals.
It shows that male customers give a higher average tip than female customers.
